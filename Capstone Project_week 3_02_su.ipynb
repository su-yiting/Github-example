```python
# importing the libraries
import geocoder
import pandas as pd
```

```python
# Reading the csv file create by the part 1
df = pd.read_csv("torontc.csv")
df.head()
```

```python
# get the coordinate 
def get_cor(postal_code):
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Toronto, Ontario'.format(postal_code))
        lat_lng_coords = g.latlng
    return lat_lng_coords

# get the all PostalCode Coordinates
postal_codes = df['PostalCode']    
cor = [get_cor(postal_code) for postal_code in postal_codes.tolist()]
```

```python
# Add Latitude and Longitude columns
df_cor = pd.DataFrame(cor, columns=['Latitude', 'Longitude'])
df['Latitude'] = df_cor['Latitude']
df['Longitude'] = df_cor['Longitude']
```
